```python
import csv
import random

file_list = ['NACC_ad.csv', 'NACC_mci.csv', 'NACC_normal.csv']
merged_data = []
header = []

for i, file_name in enumerate(file_list):
    with open(file_name, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        current_header = next(reader)
        
        if i == 0:
            header = current_header
            
        for row in reader:
            merged_data.append(row)

random.seed(42)
random.shuffle(merged_data)

final_data = [header] + merged_data

with open('NACC诊断.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(final_data)

print("数据合并完成，已保存为 NACC诊断.csv")
```
