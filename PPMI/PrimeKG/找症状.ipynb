```python
import pandas as pd

# 你的文件路径
FILE_PATH = "primekg_parkinson_only.csv"

def main():
    print(f"正在读取 {FILE_PATH} ...")
    df = pd.read_csv(FILE_PATH, low_memory=False)
    
    # 自动识别列名 (防止 head/x_name 混淆)
    cols = df.columns
    x_type_col = 'x_type'
    y_type_col = 'y_type'
    x_name_col = 'x_name' if 'x_name' in cols else 'head'
    y_name_col = 'y_name' if 'y_name' in cols else 'tail'

    # 提取所有类型为 'effect/phenotype' 的节点名字
    phenotypes = set()
    
    # 检查 x 列是不是症状
    p_x = df[df[x_type_col] == 'effect/phenotype'][x_name_col]
    phenotypes.update(p_x.tolist())
    
    # 检查 y 列是不是症状
    p_y = df[df[y_type_col] == 'effect/phenotype'][y_name_col]
    phenotypes.update(p_y.tolist())
    
    # 排序打印，方便你复制
    print("\n" + "="*50)
    print(f"📋 发现 {len(phenotypes)} 个可用症状节点 (请从中挑选填入 SYMPTOM_MAP)")
    print("="*50)
    
    for p in sorted(list(phenotypes)):
        print(f"'{p}'")
        
    print("="*50)

if __name__ == "__main__":
    main()
```
