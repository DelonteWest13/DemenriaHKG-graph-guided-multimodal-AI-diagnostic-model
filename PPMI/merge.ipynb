```python
import csv
import random

file_list = ['PD1.csv', 'prodromal.csv', 'swedd.csv', 'control.csv']
merged_data = []
header = []

for i, file_name in enumerate(file_list):
    with open(file_name, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        current_header = next(reader)
        
        if i == 0:
            header = current_header
            
        for row in reader:
            merged_data.append(row)

random.seed(42)
random.shuffle(merged_data)

final_data = [header] + merged_data

with open('PPMI诊断.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(final_data)

print("数据合并完成，已保存为 PPMI诊断.csv")
```

```python
import csv
import os
import random

file_folder_map = [
    ('PD1.csv', 'PD/'),
    ('control.csv', 'Control/'),
    ('swedd.csv', 'SWEDD/'),
    ('prodromal.csv', 'Prodromal/')
]

merged_data = []
header = []
total_valid_count = 0
total_deleted_count = 0

for file_name, folder_name in file_folder_map:
    valid_data = []
    deleted_count = 0
    valid_count = 0
    
    with open(file_name, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        current_header = next(reader)
        
        if not header:
            header = current_header
            
        valid_data.append(current_header)
        
        for row in reader:
            path = folder_name + row[1]
            if os.path.exists(path):
                valid_data.append(row)
                merged_data.append(row)
                valid_count += 1
            else:
                deleted_count += 1
                
    with open(file_name, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerows(valid_data)
        
    print(file_name + f" 检查完毕。保留数据：{valid_count}条，删除无效数据：{deleted_count}条。")
    total_valid_count += valid_count
    total_deleted_count += deleted_count

random.seed(42)
random.shuffle(merged_data)

final_data = [header] + merged_data

with open('PPMI诊断.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(final_data)

print(f"最终数据已保存为 PPMI诊断.csv。总计保留数据：{total_valid_count}条，总计删除无效数据：{total_deleted_count}条。")
```
