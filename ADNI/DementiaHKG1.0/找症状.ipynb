```python
import pandas as pd
import os

# 指向刚才生成的 AD 子图
FILE_PATH = "primekg_ad_only.csv"

def main():
    print(f"🕵️ 正在体检知识库: {FILE_PATH} ...")
    
    if not os.path.exists(FILE_PATH):
        print(f"❌ 错误：找不到文件 {FILE_PATH}，请先运行 extract_ad_primekg_subgraph.py")
        return

    df = pd.read_csv(FILE_PATH, low_memory=False)

    # 自动识别列名
    cols = df.columns
    x_type_col = 'x_type'
    y_type_col = 'y_type'
    x_name_col = 'x_name' if 'x_name' in cols else 'head'
    y_name_col = 'y_name' if 'y_name' in cols else 'tail'

    # ================= 1. 检查症状/表型 (用于临床评分映射) =================
    phenotypes = set()
    
    # 提取两端类型为 effect/phenotype 的节点
    p_x = df[df[x_type_col] == 'effect/phenotype'][x_name_col]
    p_y = df[df[y_type_col] == 'effect/phenotype'][y_name_col]
    
    phenotypes.update(p_x.dropna().tolist())
    phenotypes.update(p_y.dropna().tolist())

    print("\n" + "="*60)
    print(f"🧠 [症状/表型] 检查报告: 共发现 {len(phenotypes)} 个节点")
    print("   👉 作用：用于映射 ADNI 的临床表格 (DX, MMSE, CDR 等)")
    print("="*60)
    
    # 排序打印，重点找 Dementia, Memory, Cognitive
    for p in sorted(list(phenotypes)):
        print(f"  - '{p}'")

    # ================= 2. 检查核心基因 (用于生物标志物映射) =================
    genes = set()
    g_x = df[df[x_type_col] == 'gene/protein'][x_name_col]
    g_y = df[df[y_type_col] == 'gene/protein'][y_name_col]
    genes.update(g_x.dropna().tolist())
    genes.update(g_y.dropna().tolist())

    print("\n" + "="*60)
    print(f"🧬 [基因/蛋白] 检查报告: 共发现 {len(genes)} 个节点")
    print("   👉 作用：确认图谱是否包含 AD 核心病理 (Amyloid/Tau/APOE)")
    print("="*60)

    # 自动高亮核心基因，方便你确认
    core_targets = ['APOE', 'APP', 'MAPT', 'PSEN1', 'PSEN2', 'TREM2', 'ABCA7']
    found_cores = []
    
    for g in sorted(list(genes)):
        # 如果是核心基因，加入列表
        if g in core_targets:
            found_cores.append(g)
            
    print(f"✅ 核心基因命中情况: {found_cores}")
    print(f"   (APOE=载脂蛋白E, APP=淀粉样前体, MAPT=Tau蛋白, PSEN=早老素)")
    
    if len(found_cores) < 3:
        print("⚠️ 警告：核心基因缺失较多，图谱的生物学深度可能不够！")
    else:
        print("🎉 完美：核心致病基因都在，图谱质量很高！")

if __name__ == "__main__":
    main()
```
