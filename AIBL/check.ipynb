```python
import torch
from thop import profile
import model1

def evaluate_models():
    results = []
    
    # CNN_3D
    model_cnn = model1.CNN_3D(num_class=1)
    input_cnn = torch.randn(1, 1, 64, 64, 64)
    macs, params = profile(model_cnn, inputs=(input_cnn,), verbose=False)
    results.append(("CNN_3D", params, macs))
    
    # NeuralNet
    model_nn = model1.NeuralNet(embedding=128)
    input_nn = torch.randn(1, 128)
    macs, params = profile(model_nn, inputs=(input_nn,), verbose=False)
    results.append(("NeuralNet", params, macs))
    
    # DualTransformer
    model_dual = model1.DualTransformer()
    input_dual = torch.randn(1, 2)
    macs, params = profile(model_dual, inputs=(input_dual,), verbose=False)
    results.append(("DualTransformer", params, macs))
    
    # MultiModalTransformer
    model_multi = model1.MultiModalTransformer()
    input_multi = torch.randn(1, 3)
    macs, params = profile(model_multi, inputs=(input_multi,), verbose=False)
    results.append(("MultiModalTransformer", params, macs))
    
    # KGMultiModalTransformer_old
    model_kg_old = model1.KGMultiModalTransformer_old()
    input_kg_old_x = torch.randn(1, 3)
    input_kg_old_transe = torch.randn(1, 32)
    macs, params = profile(model_kg_old, inputs=(input_kg_old_x, input_kg_old_transe), verbose=False)
    results.append(("KGMultiModalTransformer_old", params, macs))
    
    # KGMultiModalTransformer
    model_kg = model1.KGMultiModalTransformer(ehr_dim=2, img_dim=64, bio_dim=5, max_seq_len=20)
    input_kg_x = torch.randn(1, 71) 
    input_kg_seq = torch.randn(1, 20, 128)
    input_kg_mask = torch.zeros(1, 20)
    macs, params = profile(model_kg, inputs=(input_kg_x, input_kg_seq, input_kg_mask), verbose=False)
    results.append(("KGMultiModalTransformer", params, macs))
    
    # KGMultiModalPerceiver
    model_perceiver = model1.KGMultiModalPerceiver()
    input_perceiver_x = torch.randn(1, 3)
    input_perceiver_transe = torch.randn(1, 32)
    macs, params = profile(model_perceiver, inputs=(input_perceiver_x, input_perceiver_transe), verbose=False)
    results.append(("KGMultiModalPerceiver", params, macs))
    
    print(f"{'Method Name':<30} | {'Parameters':<15} | {'Compute Cost (MACs)':<20}")
    print("-" * 70)
    for name, params, macs in results:
        print(f"{name:<30} | {params:<15.0f} | {macs:<20.0f}")

if __name__ == "__main__":
    evaluate_models()
```
